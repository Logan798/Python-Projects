```python
import pylab as plt

#Create a bunch of lists for the purpose of plotting
samples = []
linear = []
quad = []
cube = []
expon = []

for i in range (0,30):
    samples.append(i)
    linear.append(i)
    quad.append(i**2)
    cube.append(i**3)
    expon.append(1.5**i)

plt.plot(samples,linear)#Takes x and y arguments
plt.plot(samples,quad)
plt.plot(samples,cube)
plt.plot(samples,expon)
```

## Using figure() to separate the graphs for a better view
## Takes a string for the name of the plot as an argument

```python
plt.figure("Linear")
plt.plot(samples,linear)
plt.figure("Quadratic")
plt.plot(samples,quad)
plt.figure("Cubic")
plt.plot(samples,cube)
plt.figure("Exponential")
plt.plot(samples,expon)
```

## Giving labels to the axises, and giving titles to the graphs

```python
plt.figure("Linear")
plt.title("Linear")
plt.xlabel("Sample Points")
plt.ylabel("Linear Function")
plt.plot(samples,linear)
plt.figure("Quadratic")
plt.title("Quadratic")
plt.plot(samples,quad)
plt.figure("Cubic")
plt.title("Cubic")
plt.plot(samples,cube)
plt.figure("Exponential")
plt.title("Exponential")
plt.plot(samples,expon)
#Lets you add just a label to one axis
plt.figure("Quadratic")
plt.ylabel("Quadratic function")
```

## Using clf()(Clear figure)

```python
plt.figure("Linear")
plt.clf()
plt.plot(samples,linear)
plt.figure("Quadratic")
plt.clf()
plt.plot(samples,quad)
plt.figure("Cubic")
plt.clf()
plt.plot(samples,cube)
plt.figure("Exponential")
plt.clf()
plt.plot(samples,expon)

plt.figure("Linear")
plt.title("Linear")
plt.figure("Quadratic")
plt.title("Quadratic")
plt.figure("Cubic")
plt.title("Cubic")
plt.figure("Exponential")
plt.title("Exponential")
```

## Comparing graphs

```python
plt.figure("Linear")
plt.clf()
plt.title("Linear")
plt.ylim(0,1000)#Sets the bounds for the y-axis
plt.plot(samples,linear)
plt.figure("Quadratic")
plt.clf()
plt.title("Quadratic")
plt.ylim(0,1000)
plt.plot(samples,quad)
```

## Overlaying Plots & Using labels and a legend

```python
plt.figure("Linear Quad")
plt.clf()
plt.title("Linear vs Quadratic")
#Plots and adds a label to each line
plt.plot(samples,linear,label = 'Linear')
plt.plot(samples,quad, label = 'Quadratic')
plt.legend(loc = 'upper left')#Can specify a location for the legend

plt.figure("Cubic Expon")
plt.clf()
plt.title("Cubic vs Exponential")
plt.plot(samples,cube,label = "Cubic")
plt.plot(samples,expon,label = "Exponential")
plt.legend()#Defaults to what it thinks it is best to place it
```

## Changing data display

```python
#The little string in plot, like 'b-' specify color
plt.figure("Linear Quad")
plt.clf()
plt.title("Linear vs Quadratic")
#Plots and adds a label to each line
plt.plot(samples,linear,'b-',label = 'Linear',linewidth = 2.0)
plt.plot(samples,quad,'ro', label = 'Quadratic',linewidth = 3.0)
plt.legend(loc = 'upper left')#Can specify a location for the legend

plt.figure("Cubic Expon")
plt.clf()
plt.title("Cubic vs Exponential")
plt.plot(samples,cube,'g^',label = "Cubic", linewidth = 4.0)
plt.plot(samples,expon,'r--',label = "Exponential",linewidth = 5.0)
plt.legend()#Defaults to what it thinks it is best to place it
```

## Using subplots

```python
#The little string in plot, like 'b-' specify color
plt.figure("Linear Quad")
plt.clf()
plt.subplot(211)#Argument is rows, columets, and location; For instance, '211' says two rows, one columun, and the location to use
plt.ylim(0,900)
plt.title("Linear vs Quadratic")
#Plots and adds a label to each line
plt.plot(samples,linear,'b-',label = 'Linear',linewidth = 2.0)
plt.subplot(212)
plt.ylim(0,900)
plt.plot(samples,quad,'ro', label = 'Quadratic',linewidth = 3.0)
plt.legend(loc = 'upper left')#Can specify a location for the legend

plt.figure("Cubic Expon")
plt.clf()
plt.subplot(121)
plt.ylim(0,140000)
plt.title("Cubic vs Exponential")
plt.plot(samples,cube,'g^',label = "Cubic", linewidth = 4.0)
plt.subplot(122)
plt.ylim(0,140000)
plt.plot(samples,expon,'r--',label = "Exponential",linewidth = 5.0)
plt.legend()#Defaults to what it thinks it is best to place it
```

## Changing Scales

```python
plt.figure("cube exp log")
plt.clf()
plt.plot(samples,cube,'g--',label = 'Cubic',linewidth = 2.0)
plt.plot(samples,expon,'r',label = 'Exponential',linewidth = 4.0)
plt.yscale('log')
plt.legend()
plt.title("Cubic vs Exponential")

plt.figure("cube exp linear")
plt.clf()
plt.plot(samples,cube,'g--',label = 'Cubic',linewidth = 2.0)
plt.plot(samples,expon,'r',label = 'Exponential',linewidth = 4.0)
plt.legend()
plt.title("Cubic vs Exponential")
```
