```python
# Hangman game

# -----------------------------------
# Helper code
# You don't need to understand this helper code,
# but you will have to know how to use the functions
# (so be sure to read the docstrings!)

import random
import string#For a lowercase string of the English Alphabet

WORDLIST_FILENAME = "words.txt"

def loadWords():
    """
    Returns a list of valid words. Words are strings of lowercase letters.
    
    Depending on the size of the word list, this function may
    take a while to finish.
    """
    print("Loading word list from file...")
    # inFile: file
    inFile = open(WORDLIST_FILENAME, 'r')
    # line: string
    line = inFile.readline()
    # wordlist: list of strings
    wordlist = line.split()
    print("  ", len(wordlist), "words loaded.")
    return wordlist

def chooseWord(wordlist):
    """
    wordlist (list): list of words (strings)

    Returns a word from wordlist at random
    """
    return random.choice(wordlist)

# end of helper code
# -----------------------------------

# Load the list of words into the variable wordlist
# so that it can be accessed from anywhere in the program
wordlist = loadWords()

def isWordGuessed(secretWord, lettersGuessed):
    '''
    secretWord: string, the word the user is guessing
    lettersGuessed: list, what letters have been guessed so far
    returns: boolean, True if all the letters of secretWord are in lettersGuessed;
      False otherwise
    '''
    # FILL IN YOUR CODE HERE...
    #Created by: Logan Crockett
    #On: 9/21/19
    #CPSC 2100
    for letter in secretWord:#For each letter in the secret word
        if letter not in lettersGuessed:
            return False#If it is not the the list of guesses, return false
        else:
            continue#Otherwise, continue testing
    #If it reaches here, then all letters guessed are in the word
    return True
#Test if the function 'isWordGuessed' here
#Comment out for individual testing
#word = 'apple'
#lettersGuessed = ['a','c','d','e','l','m','p']
#print(isWordGuessed(word,lettersGuessed))


def getGuessedWord(secretWord, lettersGuessed):
    '''
    secretWord: string, the word the user is guessing
    lettersGuessed: list, what letters have been guessed so far
    returns: string, comprised of letters and underscores that represents
      what letters in secretWord have been guessed so far.
    '''
    # FILL IN YOUR CODE HERE...
    #Created by: Logan Crockett
    #On: 9/21/19
    #CPSC 2100
    guessedWord = ''#Start with a blank string
    for letter in secretWord:#Test for which letters have been correctly guessed
        if letter in lettersGuessed:#If it is in the word, add it to the string
            guessedWord = guessedWord + letter
        else:#If it is not, then add an underscore
            guessedWord = guessedWord + "_ "
    #When finished, return the string
    return guessedWord
#Test for the function 'getGuessedWord()'
#Comment out to test the function
#word = 'apple'
#lettersGuessed = ['o','c','d','m','d']
#print("Word: " + getGuessedWord(word,lettersGuessed))


def getAvailableLetters(lettersGuessed):
    '''
    lettersGuessed: list, what letters have been guessed so far
    returns: string, comprised of letters that represents what letters have not
      yet been guessed.
    '''
    # FILL IN YOUR CODE HERE...
    #Created by: Logan Crockett
    #On: 9/21/19
    #CPSC 2100
    allLetters = string.ascii_lowercase#Start with the string of lowercase letters
    availableLetters = ''#Start with a blank string
    for letter in allLetters:#For each letter, do not add it into the list
        if letter not in lettersGuessed:#If it is not in the guesses, add it to the string
            availableLetters += letter
        #Otherwise, continue testing
    #Return the string of available letters
    return availableLetters
#Test for the function 'getAvailableLetters()'
#Comment out for testing
#guesses = ['a','g','l','o','r','z']
#print("Available letters: " + getAvailableLetters(guesses))

def hangman(secretWord):
    '''
    secretWord: string, the secret word to guess.

    Starts up an interactive game of Hangman.

    * At the start of the game, let the user know how many 
      letters the secretWord contains.

    * Ask the user to supply one guess (i.e. letter) per round.

    * The user should receive feedback immediately after each guess 
      about whether their guess appears in the computers word.

    * After each round, you should also display to the user the 
      partially guessed word so far, as well as letters that the 
      user has not yet guessed.

    Follows the other limitations detailed in the problem write-up.
    '''
    # FILL IN YOUR CODE HERE...
    #Created by: Logan Crockett
    #On: 9/21/19
    #CPSC 2100
    lettersGuessed = list()#List of letters that have been guessed
    guesses = 8#Number of guesses allowed
    
    #Print a generic block denoting the game
    print("Welcome to a game called Hangman.")
    print("I am thinking of a word that is " + str(len(secretWord)) + " letters long.")
    print("----------")
    
    #Run the code while the word is not guessed
    while guesses > 0:#Zero indicates they ran out of guesses
        print("You have " + str(guesses) + " guesses left.")
        print("Available letters: " + getAvailableLetters(lettersGuessed))
        userGuess = input("Your guess: ")
        guessInLower = userGuess.lower()#Ensure the guess is in lowercase
        
        #Check for duplicate input
        if guessInLower in lettersGuessed:
            print("Sorry. You have already guessed that. Please guess again.")
            print("----------")
        else:#Otherwise, check if it is in the word
            lettersGuessed.append(guessInLower)#Add the word to the guesses
            
            if guessInLower in secretWord:#If the guess is in the word, update the word
                print("Good Guess: " + getGuessedWord(secretWord,lettersGuessed))
                print("---------")
                #Do not subtract from the guesses if they guessed correctly
            else:#If it is not, update the amount of guesses left
                print("Oops! That guess wasn't correct: " + getGuessedWord(secretWord,lettersGuessed))
                print("---------")
                guesses -= 1#Subtract by one guesses
        
            #Test if the word has finally been guessed
            if isWordGuessed(secretWord,lettersGuessed):
            #If so, then tell user, and stop game
                print("Congratulations! You win.")
                return
            #If not, then continue through the loop  
    
    
    #If they do not correctly guess through the loop, then inform user and display the word
    print("Sorry. The word was " + secretWord)
    return





# When you've completed your hangman function, uncomment these two lines
# and run this file to test! (hint: you might want to pick your own
# secretWord while you're testing)

secretWord = chooseWord(wordlist).lower()
hangman(secretWord)
```
